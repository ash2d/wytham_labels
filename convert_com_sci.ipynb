```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df_counts = pd.read_csv('species_frequency_counts.csv')
df_counts.head()
```

```python
df_conv = pd.read_csv('common_sci.csv')
df_conv = df_conv[['PRIMARY_COM_NAME','SCI_NAME']]
df_conv.head()
```

```python
df_conv_freq = df_counts.merge(df_conv, left_on='Species', right_on='SCI_NAME', how='left').drop(columns=['SCI_NAME'])
df_conv_freq.head(12)
```

```python
df_conv_freq.head(12)
```

```python
# count records with no PRIMARY_COM_NAME
missing_count = df_conv_freq['PRIMARY_COM_NAME'].isna().sum()
print(missing_count)
```

```python
unconv = df_conv_freq[df_conv_freq['PRIMARY_COM_NAME'].isna()]
print("Unconverted species:")
unconv.head()
```

```python
nilo_freq = [1646370, 1592189, 929896, 829954, 460416, 399303, 327395, 138739, 129551, 121672]
nilo_spec = ['Eurasian Wren', 'European Robin', 'Song Thrush', 'Eurasian Blackbird', 'Common Chiffchaff', 'Eurasian Blue Tit', 'Eurasian Blackcap', 'Coal Tit', 'Tawny Owl', 'Eurasian Nuthatch']
```

```python
df_nilo = pd.DataFrame({'Species': nilo_spec, 'Frequency': nilo_freq})
df_nilo.head()
```

```python
df_compare = df_conv_freq.merge(df_nilo, left_on='PRIMARY_COM_NAME', right_on='Species', how='left')
df_compare['ratio'] = df_compare['Frequency_y'] / df_compare['Frequency_x']
df_compare.head(15)
```
